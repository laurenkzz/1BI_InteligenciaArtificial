```python
# Exercício Aula 01 - Bibliotecas.

# Pandas  é utilizada para manipulação de dados. NumPy é utilizada para cálculos numéricos rápidos. Matplotlib é utilizada para criar gráficos e visualizações de dados.

import pandas as pd

# Criando um "dataset" de livros
dados = {
    "titulo": ["Livro A", "Livro B", "Livro C"],
    "autor": ["Autor 1", "Autor 2", "Autor 3"],
    "paginas": [200, 350, 150],
    "vendas": [5000, 12000, 8000]
}

df = pd.DataFrame(dados)

# Mostrar tabela
print(df)

# Filtrar livros
mais_vendidos = df[df["vendas"] > 8000]

print("\nLivros mais vendidos:")
print(mais_vendidos)

import matplotlib.pyplot as plt

titulos = ["Livro A", "Livro B", "Livro C"]
vendas = [5000, 12000, 8000]

# Gráfico de barras
plt.bar(titulos, vendas)

plt.title("Vendas por Livro")
plt.xlabel("Livros")
plt.ylabel("Quantidade de Vendas")

plt.show()

import matplotlib.pyplot as plt

titulos = ["Livro A", "Livro B", "Livro C"]
vendas = [5000, 12000, 8000]

# Gráfico de barras
plt.bar(titulos, vendas)

plt.title("Vendas por Livro")
plt.xlabel("Livros")
plt.ylabel("Quantidade de Vendas")

plt.show()

```
